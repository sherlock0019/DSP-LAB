```python
#Question - 6

#slow convolution

def convolution(x,y):
    x = np.array(x)
    y = np.array(y)
    
    N1 = len(x)
    N2 = len(y)
    N = N1+N2-1
    conv = np.zeros(N)
    a = N-N1
    b = N-N2
    
    #Padding zeros to x and h to make their length to N
    x =np.pad(x,(0,a),'constant')
    y =np.pad(y,(0,b),'constant')
    
    #Linear convolution using convolution sum formula
    for n in np.arange(0,N,1):
        for k in np.arange(0,N,1):
            if n >= k:
                conv[n] = conv[n]+x[n-k]*y[k]
    
    return conv
```

```python
#DTFT and IDTFT
#fast convolution

def DTFT(x):
    x = np.array(x)
    w = np.linspace(0, 2*np.pi, 100)
    x_ejw = []

    for i in range(len(w)):
        x_ejw.append(np.sum(x*np.exp(-1j * w[i]* np.arange(len(x)))))

    magnitude_spec = []

    for i in range(len(w)):
        magnitude_spec.append(np.linalg.norm(x_ejw[i]))

    plt.title('Magnitude specturm of the DTFT')
    plt.plot(w, magnitude_spec)
    plt.xlabel('Frequency')
    plt.show()

    phase_spec = []

    for i in range(len(w)):
        phase_spec.append(cmath.phase(x_ejw[i]))

    plt.title('Phase specturm of the DTFT')
    plt.plot(w, phase_spec)
    plt.xlabel('Frequency')
    plt.show()
    
    return x_ejw, magnitude_spec, phase_spec
```

```python
#inverse Discrete fourier transform

def IDTFT(x, n_samples):
    x = np.array(x)
    w = np.linspace(0, 2*np.pi, 100)
    n= np.linspace(0, n_samples, 1)
    x_n = []

    for i in range(n_samples):
        x_n.append(np.sum(x*np.exp(1j*w*i)))

    x_n = np.array(x_n)
    x_n = np.real((x_n/100))
    
    return x_n
```

```python
#Question - 7
#4 point DFT for x[n]

def DFT(x):
    """
    Function to calculate the 
    discrete Fourier Transform 
    of a 1D real-valued signal x
    """
    x = np.array(x)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1)) #taking transpose
    e = np.exp(-2j * np.pi * k * n / N)
    
    X_k = np.dot(e, x)
    
    return X_k

def IDFT(x):
    """
    Function to calculate the 
    inverse discrete Fourier Transform 
    of a 1D real-valued signal x
    """
    x = np.array(x)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1)) #taking transpose
    e = np.exp(2j * np.pi * k * n / N)
    
    X_n = (1/N)*np.dot(e, x)
    
    return X_n

#input signal is
s_n = np.array([3, 1, 2, 1])
s_k = DFT(s_n)

print('4 point DFT of the signal is ', s_k)

#comparision between the inbuilt function
m = sp.linalg.dft(s_n.shape[0])
print('4 point DFT of the inbuilt function', m @ s_n)

#checking with reconstruction
recon_signal = np.real(IDFT(s_k))
print('Reconstructed signal is ', recon_signal)
```

```python
#Circular convolution

def rotate(A):
    n = len(A)
    
    x = A[n - 1]
    for i in range(n - 1, 0, -1):
        A[i] = A[i - 1];
    A[0] = x;
    
    return A

def CircularConvolution(h, x):
    x = np.array(x)
    h = np.array(h)
    
    #flipping array
    h = np.flip(h)
    
    conv_n = np.zeros(x.shape[0])
    
    for i in range(x.shape[0]):
        h = rotate(h)
        conv_n[i] = np.sum(x*h)
        
    return conv_n

print('The sequences are :', [2, 1, 2, 1], [1, 2, 3, 4])
print('The circular convolution is:', CircularConvolution([2, 1, 2, 1], [1, 2, 3, 4]))
```
